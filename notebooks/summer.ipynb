```ocaml
module type MON =
sig
  type t
  val one : t
  val mul : t -> t -> t
end
```

```ocaml
implicit module MON_int =
struct
  type t = int
  let one = 1
  let mul = ( * )
end
```

```ocaml
implicit module MON_intcode =
struct
  type t = int code
  let one = .<1>.
  let mul x y = .< .~x * .~y >.
end
```

```ocaml
let ( <*> ) {N:MON} x y = N.mul x y
```

```ocaml
let rec pow : {N:MON} -> N.t -> int -> N.t  =
  fun {N:MON} x n ->
    if n = 0 then N.one
    else x <*> pow x (n - 1)
```

```ocaml
let x = pow 2 3
let y = .< fun x -> .~(pow .<x>. 3) >.
```

```ocaml
type 'a sd =
    Sta : 'a -> 'a sd
  | Dyn : 'a code -> 'a sd
```

```ocaml
let sta_ s = Sta s
```

```ocaml
let dyn_ d = Dyn d
```

```ocaml
let cd_ = function
 | Sta s -> .< s >.
 | Dyn d -> d
```

```ocaml
implicit module Num_ps_int =
struct
  type t = int sd

  let int x = sta_ x

  let mul x y = match x, y with
   | Sta l  , Sta r -> Sta (l * r)
   | Sta 1, x 
   | x    , Sta 1 -> x
   | Sta 0, _
   | _    , Sta 0 -> Sta 0
   | x, y -> Dyn .< .~(cd_ x) * .~(cd_ y) >.
end
```

```ocaml
module type PS = sig
  type t
  type ps
  val dyn : ps -> t code
  val sta : t -> ps
  val cd : t code -> ps
end
```

```ocaml
let dyn {P:PS} p = P.dyn p
let sta {P:PS} s = P.sta s
let cd {P:PS} c = P.cd c
```

```ocaml
implicit module MON_int_ps =
struct
  type t = { sta: int; dyn: int code option }

  let one = { sta = 1; dyn = None }

  let mul x y = match x.sta * y.sta, x.dyn, y.dyn with
      0, _, _ -> { sta = 0; dyn = None }
    | sta, None, dyn
    | sta, dyn, None -> { sta; dyn }
    | sta, Some x, Some y -> {sta; dyn = Some (.<.~x * .~y>.)}
end
```

```ocaml
implicit module PS_int_ps (* : PS with type t = int and type ps = MON_int_ps.t *) =
struct
  open MON_int_ps
  type t = int
  type ps = MON_int_ps.t
  let cd d = { sta = 1; dyn = Some d}
  let sta s = { sta = s; dyn = None }
  let dyn = function
    | {sta=1; dyn = Some d} -> d
    | {sta; dyn = Some d} -> .< sta * .~d >.
    | {sta} -> .< sta >.
end
```

```ocaml
let pow5_sd = .< fun x -> .~(dyn (pow (cd .<x>.) 5)) >.
```
